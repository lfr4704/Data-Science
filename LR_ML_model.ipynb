# 2. Use Spacey to tokenize the total text

```python
#use spacey to tokenize the total text
import re
import string

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import spacy
from spacy.tokenizer import Tokenizer
from bs4 import BeautifulSoup
```

```python
#initialize spacy model and tokenizer
nlp = spacy.load("en_core_web_lg")
tokenizer = Tokenizer(nlp.vocab)
```

```python
tokens =[]
#return lower case tokens and ignore stop words/punctunation

for doc in tokenizer.pipe(df['all_text'], batch_size=500):
    doc_tokens=[]
    
    for token in doc:
        #ignore stopwords, spaces and punctuation
        if (token.is_stop == False) & (token.is_punct == False) & (token.is_space == False):
            doc_tokens.append(token.text.lower())
    tokens.append(doc_tokens)
df['spacy_tokens'] = tokens
df['spacy_tokens'].head()
```

```python
df.head()
```

# 3.  Vector Representation

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
```

```python
def gather_data(df_column):
    """Produces list of documents form a df column
        returns lis of strings"""
    
    #data = []
    #for row in df_column:
    #    data.append(' '.join(str(row)))
    data = [''.join(str(item)) for item in df_column]
    
    return data
```

```python
data = gather_data(df['spacy_tokens'])
data[0]
```

```python
#create the transformer
vect = CountVectorizer()

#build vocab
vect.fit(data)

#transform text
dtm = vect.transform(data)
dtm.shape
```

```python
vect.get_feature_names()
```

```python
#tunning the parameter

#instantiate vectorizer object
tfidf = TfidfVectorizer(stop_words='english',
                        ngram_range=(1,2),
                        max_df=0.97,
                        min_df=3)
#create a vocabulary and get word counts per document
dtm = tfidf.fit_transform(data)

#get feature names to use a datafame column headers
dtm = pd.DataFrame(dtm.todense(), columns=tfidf.get_feature_names())

dtm.head()
```

# 4. Train the model

```python
from sklearn.neighbors import NearestNeighbors

#fit on TF-IDF Vectors
nn = NearestNeighbors(n_neighbors=5, algorithm='kd_tree')
nn.fit(dtm)
```

```python
#get the top 5 nearest neighbors
nn.kneighbors([dtm.iloc[0].values])
```
